```python
control_head = open("./control_generator/control_head.py", "r").readlines()
control_b1= open("./control_generator/control_b1.py", "r").readlines()
control_b2 = open("./control_generator/control_b2.py", "r").readlines()
control_tail = open("./control_generator/control_tail.py", "r").readlines()

```

```python
dfLack = dfLack.replace(np.nan, '', regex=True)

for index, row in dfLack.iterrows():
    name = row['files']
    '''test
    '''
    kvalues = row['kvalue'].split(',')
    # print(kvalues)
    kvalues = [s.strip() for s in kvalues]

    head_code = control_head
    kvalues_code = [f'\nkp = {kvalues[0]}\nkd = {kvalues[1]}\nkp_vel = {kvalues[2]}\nkd_vel = {kvalues[3]}\n']
    b1_code = control_b1
    b2_code = control_b2
    tail_code = control_tail

    if row['servo'].strip() != '':
      b1_code = [l.replace('servo_offset = 18.0*math.pi/180', row['servo'].strip()) for l in b1_code]

    if row['if_velocity'].strip() == 'comment if velocity':
       velocity_code = ['\n        if velocity > 2.5:\n                velocity = 2*velocity\n']
    elif row['if_velocity'].strip() == 'remove all velocity constraints ( remove velocity if statements)':
      velocity_code = ['']
    elif row['if_velocity'].strip() == 'if velocity > 2.5: velocity = 2*velocity':
      velocity_code = ['\n        if velocity < 0:\n                velocity = 1\n        if velocity > 2.5:\n                velocity = 2*velocity\n']
    elif row['if_velocity'].strip() == 'if velocity > 2.5: velocity = velocity':
      velocity_code = ['\n        if velocity < 0:\n                velocity = 1\n        if velocity > 2.5:\n                velocity = velocity\n']
    else:
      velocity_code = ['\n        if velocity < 0:\n                velocity = 1\n        if velocity > 2.5:\n                velocity = 2.5\n']
    
    if row['velocity_control'].strip() != '':
      b2_code = [l.replace('		velocity = velocity + abs(control_error_vel)', '		' + row['velocity_control'].strip()) for l in b2_code]
    
    def if_to_value(l, x):
      if 'if' in l:
        return l.replace('180', x)
      else:
        return l

    if row['if_angle'].strip() == 'angle if statement 180 to 0.18':
      b2_code = [if_to_value(l, '0.18') for l in b2_code]
    elif row['if_angle'].strip() == 'angle if statement 180 to 0.19':
      b2_code = [if_to_value(l, '0.19') for l in b2_code]

    if row['angle'].strip() != '':
      angle_code = ['\n		' + row['angle'].strip() + '\n']
    else:
      angle_code = ['\n		' + 'angle = angle + control_error*np.pi/180' + '\n']
    b2_code = b2_code[1:]
    code = head_code + kvalues_code + b1_code + angle_code + b2_code + velocity_code + tail_code
    with open(f"./control_py/{name}.py", 'w') as writer:
      writer.writelines(code)
      
    
```
