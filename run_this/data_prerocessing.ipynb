```python
def modify_servo(df, index):
  listOptions = ["servo_offset = 9.0*math.pi/180", "servo_offset = 36.0*math.pi/180", "servo_offset = 18.0*math.pi/180", "servo_offset = 18.0*math.pi/363"]
  optionSelect = random.randint(0,len(listOptions)-1)
  df.iloc[[index],[dictColumn['servo']]] = listOptions[optionSelect]
def modify_if_velocity(df, index):
  listOptions = ["if velocity > 2.5: velocity = velocity", "remove all velocity constraints ( remove velocity if statements)", "comment if velocity"]
  optionSelect = random.randint(0,len(listOptions)-1)
  df.iloc[[index],[dictColumn['if_velocity']]] = listOptions[optionSelect]
def modify_velocity_control(df, index):
  listOptions = ["velocity = velocity + 2*abs(control_error_vel)"]
  optionSelect = random.randint(0,len(listOptions)-1)
  df.iloc[[index],[dictColumn['velocity_control']]] = listOptions[optionSelect]
def modify_if_angle(df, index):
  listOptions = ["angle if statement 180 to 0.18", "angle if statement 180 to 0.19"]
  optionSelect = random.randint(0,len(listOptions)-1)
  df.iloc[[index],[dictColumn['if_angle']]] = listOptions[optionSelect]
def modify_angle(df, index):
  listOptions = ["angle = angle + 2*control_error*np.pi/180", "angle = 10*angle + control_error*np.pi/180"]
  optionSelect = random.randint(0,len(listOptions)-1)
  df.iloc[[index],[dictColumn['angle']]] = listOptions[optionSelect]
```
